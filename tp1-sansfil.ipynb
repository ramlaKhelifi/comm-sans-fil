##### <p style="color:#191970">  	Afficher le spectre du signal échantillonné en utilisant <U> la fonction fft()  </U></p>

```octave
f=10;
fe=1000;
te=1/fe;
t=0:te:1;
s=sin(2*pi*f*t);
n=length(t);
sf=(1/n)*(fftshift(fft(s)));
p=linspace(-fe/2,fe/2,n);
stem(p,abs(sf));
```

<h2 style="color:#8B0000">2) Quantification uniforme  </h2>

#### <p style="color:#191970">Le pas de quantification d’un convertisseur est l’intervalle entre deux tensions successives. Lorsque le convertisseur utilise n bits pour le codage et il a une plage de mesure p (intervalle des tensions), alors le pas de quantification du convertisseur est : pas=p/2n </p>

### <p style="color:#191970">
 Ecrire un code Matlab qui permet de quantifier le signal échantillonné obtenu avec la quantification par arrondi en fixant n=3 et en tenant compte du pas calculé<br>
Représenter le signal quantifié <br>
Calculer l’erreur de quantification.
 </p>

```octave
f=10;
fe=1000;
te=1/fe;
t=0:te:1;
s=sin(2*pi*f*t);
n=length(t);
sf=(1/n)*(fftshift(fft(s)));
p=linspace(-fe/2,fe/2,n);
stem(p,abs(sf));
smin=min(s);
smax=max(s);
pas=(smax-smin)/8;
a=floor((s-smin)/pas);
b=smin*pas*a+(pas/2);
for i=1:length(b)
    if(b(i)>=smax)
        b(i)=smax-pas/2;
    end
end
plot(a);
```

<h2 style="color:#8B0000">3) Codage </h2>

### <p style="color:#191970"> 	Ecrire un code Matlab qui permet de déterminer le code binaire qui de la numérisation de signal s(t)</p>

```octave
f=10;
fe=1000;
te=1/fe;
t=0:te:1;
s=sin(2*pi*f*t);
n=length(t);
sf=(1/n)*(fftshift(fft(s)));
p=linspace(-fe/2,fe/2,n);
stem(p,abs(sf));
smin=min(s);
smax=max(s);
pas=(smax-smin)/8;
a=floor((s-smin)/pas);
b=smin*pas*a+(pas/2);
for i=1:length(b)
    if(b(i)>=smax)
        b(i)=smax-pas/2;
    end
end
num=dec2bin(a,3);
plot(t,b);
text(t,b,num);
```
